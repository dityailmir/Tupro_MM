BMI, Glukosa, Insulin dan Tekanan Darah

```python
## Cek penderita yang punya BMI, Glukosa, Insulin dan Tekanan Darah
df1_all = df1.loc[(df1['BMI'] != 0) & (df1['Insulin'] != 0) & (df1['BloodPressure'] != 0) & (df1['Glucose'] != 0)]
df1_all.shape
```

```python
# Cek penderita diabetes yang obesitas / overweight
diabetic_overweight_obese = ((df1_all ['BMI'] <= 30) & (df1_all ['Outcome'] == 1)).sum()
diabetic_overweight_obese
```

```python
# Cek BMI, Glukosa, Insulin dan Tekanan Darah penderita dengan nilai minimum
df1_all.min()
```

```python
# Cek BMI, Glukosa, Insulin dan Tekanan Darah penderita dengan nilai maksimum
df1_all.max()
```
